**Project Description**: Perform regression analysis to predict the profit of each the 50 startups mentioned in the dataset by taking all the input parameters(multi linear regression).

```python
import pandas as pd
```

```python
dafr=pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
dafr
```

```python
dafr.info()
```

```python
dafr=dafr.drop(['State'],axis=1)
dafr
```

```python
dafr.info()
```

```python
startup_numeric = dafr.select_dtypes(include=['int64','float64'])
startup_numeric  #data cleaning
#But this data already contains float values
```

```python
x=startup_numeric.iloc[:,0:3].values
y=dafr['Profit'].values
```

```python
from sklearn.model_selection import train_test_split  
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0) 
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred 
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_pred,y_test)
plt.plot(y_pred,y_test)
plt.show()
```

```python
df1 = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
df1.plot()
```

```python
df1.plot(kind = 'bar')
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test,y_pred)
```
